# Rescaling yr_renovated

```python
#A better way to work with yr_renovated would be to reinterpret that column so as we count 
#how many years before 2016 (last value in this dataframe is from 2015) did the renovation happen 
#This way a value of zero would mean no renovation and a value of (lets say) 10
#would mean renovation 10 years ago

#If no renovatios has been done we use the year the house was built instead of the renovation

#This way we have a metric that differentiates one apt built in 2015 
#from one built 1920 and renovated in 2015 in the same scale.

conditions = [df['yr_renovated']==0, df['yr_renovated']>0]

values = [2016-df['yr_built'],2016-df['yr_renovated']]

df['renovation_after_yrs'] = np.select(conditions,values)

df.head()
```

```python
# now we have a more meaningful feature
sns.scatterplot(x='renovation_after_yrs',y='price',data=df )
```

```python
df.drop('yr_renovated',axis=1,inplace=True)
```

# Now we start with our model

## Using Neural Networks

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))

model.add(Dense(19,activation='relu'))
#model.add(Dropout(0.5))

model.add(Dense(19,activation='relu'))
#model.add(Dropout(0.5))

model.add(Dense(19,activation='relu'))
#model.add(Dropout(0.5))


model.add(Dense(1,activation='relu'))

model.compile(optimizer='adam', loss='mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='loss', mode='min', patience = 50, verbose = 1)

model.fit(X_train,y_train,validation_data=(X_test,y_test), callbacks=[early_stop],
          batch_size=256, epochs=4000)
```

```python
losses = pd.DataFrame(model.history.history)
losseslog = losses.copy()
losseslog['loss'] = np.log(losses['loss'])
losseslog['val_loss'] = np.log(losses['val_loss'])
```

```python
losseslog.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,predictions)
rmse = (mse)**0.5
mae = mean_absolute_error(y_test,predictions)
r2 = r2_score(y_test,predictions)
print(f"mse = {mse}\nrmse = {rmse}\nmae = {mae}\nr2 = {r2}")
```

```python
predictions = predictions.reshape(len(predictions))
sns.scatterplot(x=predictions,y=y_test)
sns.lineplot(x=predictions,y=predictions,color='red')
```
